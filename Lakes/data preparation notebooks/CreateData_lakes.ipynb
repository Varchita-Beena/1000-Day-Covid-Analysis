```python
import lakepy as lk
import pandas as pd
```

```python
OUTPUT_PATH="data/" #path to store the csv file
```

**Total 2205 lates are there**<br><br>
The "my_lake" variable is an object of class Lake() which comes with several 1. 1. attributes<br>
2. name<br>
3. country<br>
4. continent (currently not supported for HydroWeb)<br>
5. source<br>
6. original_id<br>
7. id_No<br>
8. observation_period<br>
9. latitude (currently not supported for G-REALM)<br>
10. longitude (currently not supported for G-REALM)<br>
11. misc_data<br>
12. dataframe<br>
13. data<br><br>

This helps in creating metadata for lakes (lakes_metadata.csv) with desired attributes. It will have location, source, source, type, country, name.<br>
lakes_data.csv is made with id_no, date and water level.

```python
metadata=[]
data = pd.DataFrame(columns=['id_No', 'date', 'water_level'])
for i in range(1,2206):
    my_lake=None
    try:
        my_lake = lk.search(id_No = str(i))
    except:
        pass
    if my_lake is None:
        continue
    name=my_lake.name
    country=my_lake.country
    continent=my_lake.continent
    latitude=my_lake.latitude
    longitude=my_lake.longitude
    source=my_lake.source
    if "basin" in my_lake.metadata.columns:
        basin=my_lake.metadata['basin'].iloc[0]
    else:
        basin=""
    if "status" in my_lake.metadata.columns:
        status=my_lake.metadata['status'].iloc[0]
    else:
        status=""
    if "Type" in my_lake.metadata.columns:
        body_type=my_lake.metadata['Type'].iloc[0]
    else:
        body_type=""
    metadata.append([i,name,country,continent,latitude,longitude,source,basin,status,body_type])
    data=data.append(my_lake.data[['id_No', 'date', 'water_level']])
```

```python
data.to_csv(OUTPUT_PATH+"lakes_data.csv",index=False)
```

```python
metadata=pd.DataFrame(metadata,columns=["id","name","country","continent","lat","long","source","basin","status","type"])
metadata.to_csv(OUTPUT_PATH+"lakes_metadata.csv",index=False)
```
