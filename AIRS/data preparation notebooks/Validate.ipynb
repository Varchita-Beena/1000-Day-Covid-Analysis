All the logics for creating data at different degree resolutions can be validated here.

```python
import numpy as np
import math
import random
```

```python
places=[(-90,-180),(90,180)]
for i in range(100):
    places.append((round(random.uniform(-90,90),2),round(random.uniform(-180,180),2)))
```

```python
def do(location):
    (lat,long)=location
    #lat+=0.000001
    #long+=0.000001
    lat=min(math.ceil(float(lat)),90.0)-0.5
    long=min(math.ceil(float(long)),180.0)-0.5
    print(location,"==>",(lat,long))
```

```python
do((-90,-180))
```

```python
def do(location):
    (lat,long)=location
    #lat+=0.000001
    #long+=0.000001
    lat=round(90-(int((90-float(lat))/0.05)*0.05+0.025),3)
    long=round(180-(int((180-float(long))/0.05)*0.05+0.025),3)
    print(location,"==>",(lat,long))
```

```python
for place in places:
    do(place)
```
